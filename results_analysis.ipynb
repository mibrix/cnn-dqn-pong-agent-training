```python
import json
import matplotlib.pyplot as plt

# Load data from the JSON file
with open('log_of_performance1.json', 'r') as file:
    data = json.load(file)

# Assuming the JSON structure is like this:
# {
#     "time1": value1,
#     "time2": value2,
#     ...
# }

# Extracting the keys (times) and values (performance metrics)
times = list(data.keys())
values = list(data.values())

# Convert times to a more appropriate format if needed, for example:
# times = [int(time) for time in times] # if times are integers

# Create a line plot using matplotlib
plt.figure(figsize=(10, 6))
plt.plot(times, values, marker='o', linestyle='-', color='b')
plt.title('Performance Metrics Over Episodes')
plt.xlabel('Episode')
plt.ylabel('Mean Return Over 10 Evaluation Episodes')
plt.grid(True)
plt.show()

# Alternatively, create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(times, values, color='b')
plt.title('Performance Metrics Over Time')
plt.xlabel('Time')
plt.ylabel('Performance Metric')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
episodes = list(range(1, 11))
total_returns = [11.0, 11.0, 15.0, 2.0, 12.0, 20.0, 14.0, 9.0, 16.0, 12.0]

# Calculate mean return
mean_return = sum(total_returns) / len(total_returns)

# Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(episodes, total_returns, color='b')
plt.axhline(mean_return, color='red', linestyle='--', label=f'Mean Return: {mean_return:.2f}')
plt.xlabel('Episode')
plt.ylabel('Total Return')
plt.title('Total Return for Each Episode')
plt.legend()
plt.show()
```
